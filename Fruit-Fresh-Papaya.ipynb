```python
import glob
from kneed import KneeLocator
import cv2 as cv
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
images3=[cv.imread(file,cv.IMREAD_UNCHANGED) for file in sorted(glob.glob(r"C:\Users\USER\papaya\Papaya Black Spot\clip/*.png"))]
dir2=r"C:\Users\Arifin\Downloads\papaya\Fruit Fresh Papaya\K-Mean segmented_CLAHE"

for i,img in enumerate(images3):
    y = []
    K = range(1,15)
    for k in K:
        kmeanModel = KMeans(n_clusters=k)
        kmeanModel.fit(img)
        y.append(kmeanModel.inertia_)

    x = range(1, len(y)+1)
    kn = KneeLocator(x, y, curve='convex', direction='decreasing')
    print("Optimal k value:",kn.knee)
    kmeanModel = KMeans(n_clusters=kn.knee)
    kmeanModel.fit(img)
    
    plt.imshow(img,cmap='gray')
    plt.figure(figsize=(5,5))
        
    plt.figure(figsize=(5,5))
    plt.plot(K, y, 'bx-')
    plt.xlabel('k')
    plt.ylabel('Distortion')
    plt.title('The Elbow Method showing the optimal k')      #K-mean segmentation (CLAHE)
    plt.show()
```

```python
images3=[cv.imread(file,cv.IMREAD_UNCHANGED) for file in sorted(glob.glob(r"C:\Users\USER\papaya\Papaya Black Spot\clip/*.png"))]
import matplotlib.pyplot as plt

from skimage.feature import greycomatrix, greycoprops
from skimage import data


PATCH_SIZE = 21

# open the camera image
image = images3[4]

# select some patches from grassy areas of the image
grass_locations = [(280, 454), (342, 223), (444, 192), (455, 455)]
grass_patches = []
for loc in grass_locations:
    grass_patches.append(image[loc[0]:loc[0] + PATCH_SIZE,
                               loc[1]:loc[1] + PATCH_SIZE])

# select some patches from sky areas of the image
sky_locations = [(38, 34), (139, 28), (37, 437), (145, 379)]
sky_patches = []
for loc in sky_locations:
    sky_patches.append(image[loc[0]:loc[0] + PATCH_SIZE,
                             loc[1]:loc[1] + PATCH_SIZE])

# compute some GLCM properties each patch
xs = []
ys = []
for patch in (grass_patches + sky_patches):
    glcm = greycomatrix(patch, distances=[5], angles=[0], levels=256,
                        symmetric=True, normed=True)
    xs.append(greycoprops(glcm, 'contrast')[0, 0])
    ys.append(greycoprops(glcm, 'correlation')[0, 0])

# create the figure
fig = plt.figure(figsize=(8, 8))

# display original image with locations of patches
ax = fig.add_subplot(3, 2, 1)
ax.imshow(image, cmap=plt.cm.gray,
          vmin=0, vmax=255)
for (y, x) in grass_locations:
    ax.plot(x + PATCH_SIZE / 2, y + PATCH_SIZE / 2, 'gs')
for (y, x) in sky_locations:
    ax.plot(x + PATCH_SIZE / 2, y + PATCH_SIZE / 2, 'bs')
ax.set_xlabel('Original Image')
ax.set_xticks([])
ax.set_yticks([])
ax.axis('image')

# for each patch, plot (dissimilarity, correlation)
ax = fig.add_subplot(3, 2, 2)
ax.plot(xs[:len(grass_patches)], ys[:len(grass_patches)], 'go',
        label='Leaf')
ax.plot(xs[len(grass_patches):], ys[len(grass_patches):], 'bo',
        label='Non Candidate')
ax.set_xlabel('GLCM Dissimilarity')
ax.set_ylabel('GLCM Correlation')
ax.legend()

# display the image patches
for i, patch in enumerate(grass_patches):
    ax = fig.add_subplot(3, len(grass_patches), len(grass_patches)*1 + i + 1)
    ax.imshow(patch, cmap=plt.cm.gray,
              vmin=0, vmax=255)
    ax.set_xlabel('Leaf %d' % (i + 1))

for i, patch in enumerate(sky_patches):
    ax = fig.add_subplot(3, len(sky_patches), len(sky_patches)*2 + i + 1)
    ax.imshow(patch, cmap=plt.cm.gray,
              vmin=0, vmax=255)
    ax.set_xlabel('Non Candidate %d' % (i + 1))


# display the patches and plot
fig.suptitle('Grey level co-occurrence matrix features', fontsize=14, y=1.05)
plt.tight_layout()
plt.show()
```

```python
import glob
import pandas as pd
import numpy as np
from skimage.feature import greycomatrix, greycoprops
images4=[cv.imread(file,cv.IMREAD_UNCHANGED) for file in sorted(glob.glob(r"C:\Users\Arifin\Downloads\papaya\Fruit Fresh Papaya\K-Mean segmented_CLAHE/*.png"))]
dir4=r"C:\Users\Arifin\Downloads\papaya\Fruit Fresh Papaya\GLCM_CLAHE"
li1=[]
 
for i,image in enumerate(images4):
 
 
 
    bins = np.array([0, 16, 32, 48, 64, 80, 96, 112, 128, 144, 160, 176, 192, 208, 224, 240, 255]) #16-bit
    inds = np.digitize(image, bins)
 
    max_value = inds.max()+1
    matrix_coocurrence = greycomatrix(inds, [1], [0, np.pi/4, np.pi/2, 3*np.pi/4], levels=max_value, normed=False, symmetric=False)
 
    # GLCM properties
 
    li=[]
    contrast = greycoprops(matrix_coocurrence, 'contrast')
    contrast=contrast[0].tolist()
    homogeneity = greycoprops(matrix_coocurrence, 'homogeneity')
    homogeneity=homogeneity[0].tolist()
    energy = greycoprops(matrix_coocurrence, 'energy')
    energy=energy[0].tolist()
    correlation = greycoprops(matrix_coocurrence, 'correlation')
    correlation =correlation[0].tolist()
    for i in contrast:
        li.append(i)
    for i in homogeneity:
        li.append(i)
    for i in energy:
        li.append(i)
    for i in correlation:
        li.append(i)
    li.append('Fruit Fresh Papaya')
    print(contrast)
    print(homogeneity)
    print(energy)
    print(correlation)
 
    print(i)
    li1.append(li)
    nametable =['contrast_0', 'homogeneity_0', 'energy_0', 'correlation_0'
                ,'contrast_45', 'homogeneity_45', 'energy_45', 'correlation_45'
                ,'contrast_90', 'homogeneity_90', 'energy_90', 'correlation_90'
                ,'contrast_135', 'homogeneity_135', 'energy_135', 'correlation_135', 'label']
df=pd.DataFrame(li1, columns=nametable)
```
